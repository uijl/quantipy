```python
import pathlib

import numpy as np
import matplotlib.pyplot as plt

from quantipy import analysis

cwd = pathlib.Path.cwd()
path = cwd.joinpath("..", "data")
```

#### Initialise Data Preparation class

```python
data = analysis.DataPreparation(path)
```

#### Show available data

```python
data.summary
```

#### Obtain information on effect of steep declines

```python
df = data.prepare(percentile=0.5)
df.head(10)
```

#### Show histogram of change after 3 months

```python
plot_data = df["12W"].values
plot_data = plot_data[plot_data != "N/A"]

plt.figure(figsize = [10, 8])
plt.hist(plot_data, bins = np.linspace(-40, 80, 31))
plt.title("Histogram of changes in closing tick after 12 weeks.")
plt.ylabel("Change in percentage.")
plt.xlabel("Number of occurances per bin")
plt.show()
```
